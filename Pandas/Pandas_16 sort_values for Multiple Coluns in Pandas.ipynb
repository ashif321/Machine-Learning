```python
import pandas as pd
```

```python
nba = pd.read_csv("C:\\Users\\HCL\\Machine Learning\\csv files\\nba.csv")
nba.head()
```

```python
# Sorting multiple column in ascending order
nba.sort_values(["Team", "Name"])
```

```python
# Sorting multiple column in discending order
nba.sort_values(["Team", "Name"], ascending=False)
```

```python
# Sorting Team column in ascending order & sorting Name column in discending order
nba.sort_values(["Team", "Name"], ascending=[True,False])
```

```python
# Replace dataframe with existing dataframe
nba.sort_values(["Team", "Name"], ascending=[True,False], inplace=True)
```
